# Descripcion del Estudio

El éxito de cualquier organización radica fundamentalmente en su capacidad para atraer y retener talento humano. En este contexto, el objetivo principal de este estudio es identificar los factores que contribuyen a la retención del talento humano en una empresa y explorar los posibles elementos que podrían causar desgaste en los empleados, llevándolos a buscar oportunidades fuera de la organización.

El dataset para este análisis consta de 35 columnas y 1,470 registros. Sin embargo, para los fines de este estudio, nos centraremos principalmente en dos variables:

- Edad (Age): Variable cuantitativa.
- Campo de Educación (EducationField): Variable categórica.

```python
# Agrupar los datos por 'Attrition' y 'Gender', y contar el número de empleados en cada grupo
attrition_gender_grouped = df.groupby(['Attrition', 'Gender']).size().reset_index(name='Count')

attrition_gender_grouped
```

```python
# Que generos son los que estan mas quemados?
# Crear un gráfico de barras para mostrar el recuento de empleados por 'Attrition' y 'Gender'
plt.figure(figsize=(10, 6))
sns.barplot(x='Attrition', y='Count', hue='Gender', data=attrition_gender_grouped)

plt.title('Recuento de Empleados por Attrition y Género')
plt.xlabel('Attrition')
plt.ylabel('Recuento de Empleados')
plt.legend(title='Género')
plt.show()
```

```python
# De los que estan quemados, que rol es el que esta mas quemado?
df_attrition_yes = df[df['Attrition'] == 'Yes']

# Luego, cuenta las ocurrencias por 'JobRole'
jobrole_counts = df_attrition_yes['JobRole'].value_counts()

# Ahora, crea un gráfico de barras con esta información
plt.figure(figsize=(10, 6))
jobrole_counts.plot(kind='bar')
plt.title('Conteo de Attrition x por JobRole')
plt.xlabel('JobRole')
plt.ylabel('Conteo')
plt.xticks(rotation=55)
plt.show()
```

```python
# Boxplot de Edad por JobRole y Attrition
filtered_roles = df[df['JobRole'].isin(['Laboratory Technician', 'Sales Executive', 'Research Scientist'])]

# Ajustando el código para usar una paleta de colores más diferenciada
plt.figure(figsize=(15, 8))
sns.boxplot(x='JobRole', y='Age', hue='Attrition', data=filtered_roles, palette="Set2")

plt.title('Distribución de Edad por JobRole y Attrition')
plt.xlabel('JobRole')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.show()
```

```python
# Realizando la prueba ANOVA
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
modelo = ols('Age ~ C(EducationField)', data=df).fit()
anova_results = sm.stats.anova_lm(modelo, typ=2)
print(anova_results)
```

# PARA DESPUES

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que 'data' es tu DataFrame original
# Filtrar el conjunto de datos para incluir solo a los técnicos de laboratorio con 'Yes' en Attrition
lab_technicians = df[df['JobRole'] == 'Laboratory Technician']
lab_technicians_yes_attrition = lab_technicians[lab_technicians['Attrition'] == 'Yes']

# Crear rangos de edad
bins = [18, 25, 35, 45, 55, 60]
labels = ['18-24', '25-34', '35-44', '45-54', '55-59']
lab_technicians_yes_attrition['AgeGroup'] = pd.cut(lab_technicians_yes_attrition['Age'], bins=bins, labels=labels, right=False)

# Conteo de empleados en cada rango de edad
age_group_counts_yes_attrition = lab_technicians_yes_attrition['AgeGroup'].value_counts().sort_index()

# Crear un gráfico de barras para mostrar la distribución de empleados en los diferentes rangos de edad
plt.figure(figsize=(10, 6))
age_group_counts_yes_attrition.plot(kind='bar')
plt.title('Distribución de Edad de los Técnicos de Laboratorio con Attrition "Yes"')
plt.xlabel('Grupo de Edad')
plt.ylabel('Número de Empleados')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar los datos para incluir solo a los técnicos de laboratorio
lab_technicians = df[df['JobRole'] == 'Laboratory Technician']

# Crear un Boxplot para visualizar la distribución de años en la empresa por nivel de involucración laboral
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobInvolvement', y='YearsAtCompany', data=lab_technicians)

plt.title('Años en la Empresa por Nivel de Involucración Laboral de los Técnicos de Laboratorio')
plt.xlabel('Nivel de Involucración Laboral (JobInvolvement)')
plt.ylabel('Años en la Empresa (YearsAtCompany)')
plt.xticks(rotation=45)
plt.show()
```
